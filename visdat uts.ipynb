#### SUVs have higher average selling prices than sedans/trucks - Testing the hypothesis

```python
# First, let's check what body types are available in the dataset
print("Available body types in the dataset:")
print(df_filled_no_outlier['body'].value_counts())
print(f"\nTotal unique body types: {df_filled_no_outlier['body'].nunique()}")
```

```python
# Clean and standardize body types
df_body_analysis = df_filled_no_outlier.copy()
df_body_analysis['body_clean'] = df_body_analysis['body'].str.lower().str.strip()

# Categorize into main groups: SUV, Sedan, Truck, and Others
def categorize_body_type(body):
    body = str(body).lower()
    if 'suv' in body or 'crossover' in body:
        return 'SUV'
    elif 'sedan' in body:
        return 'Sedan'
    elif 'truck' in body or 'pickup' in body:
        return 'Truck'
    elif 'coupe' in body:
        return 'Coupe'
    elif 'wagon' in body:
        return 'Wagon'
    elif 'van' in body or 'minivan' in body:
        return 'Van'
    elif 'hatchback' in body:
        return 'Hatchback'
    else:
        return 'Other'

df_body_analysis['body_category'] = df_body_analysis['body_clean'].apply(categorize_body_type)

# Calculate average selling price by body type
body_price_stats = df_body_analysis.groupby('body_category').agg({
    'sellingprice': ['mean', 'median', 'std', 'count']
}).reset_index()

body_price_stats.columns = ['body_type', 'mean_price', 'median_price', 'std_price', 'count']
body_price_stats = body_price_stats.sort_values('mean_price', ascending=False)

print("\nBody Type Price Statistics:")
print("="*80)
for _, row in body_price_stats.iterrows():
    print(f"{row['body_type']:<15} | Avg: ${row['mean_price']:>10,.2f} | Median: ${row['median_price']:>10,.2f} | Count: {int(row['count']):>7,}")
print("="*80)
```

```python
# Create an elegant visualization comparing body types
# Filter to show main categories (excluding "Other" for clearer comparison)
body_price_main = body_price_stats[body_price_stats['body_type'] != 'Other'].copy()

# Create the figure
fig = go.Figure()

# Define colors for each body type with SUV, Sedan, and Truck highlighted
color_map = {
    'SUV': '#28a745',      # Green - highlight
    'Sedan': '#dc3545',    # Red - highlight
    'Truck': '#ffc107',    # Amber - highlight (but no Truck in data)
    'Coupe': '#6c757d',
    'Van': '#6c757d',
    'Wagon': '#6c757d',
    'Hatchback': '#6c757d'
}

colors = [color_map.get(body, '#6c757d') for body in body_price_main['body_type']]

# Add bar chart
fig.add_trace(go.Bar(
    x=body_price_main['body_type'],
    y=body_price_main['mean_price'],
    marker=dict(
        color=colors,
        line=dict(color='white', width=2),
        opacity=0.9
    ),
    text=body_price_main['mean_price'],
    texttemplate='$%{text:,.0f}',
    textposition='outside',
    textfont=dict(size=13, color='#2c3e50', family='Arial Bold'),
    hovertemplate='<b>%{x}</b><br>' +
                  'Average Price: $%{y:,.2f}<br>' +
                  'Vehicles: %{customdata:,}<br>' +
                  '<extra></extra>',
    customdata=body_price_main['count']
))

# Add a horizontal line for SUV average to compare
suv_avg = body_price_main[body_price_main['body_type'] == 'SUV']['mean_price'].values[0]
sedan_avg = body_price_main[body_price_main['body_type'] == 'Sedan']['mean_price'].values[0]

fig.add_hline(
    y=suv_avg,
    line_dash="dash",
    line_color="#28a745",
    line_width=2,
    annotation_text=f"SUV Average: ${suv_avg:,.0f}",
    annotation_position="right",
    annotation_font=dict(size=11, color='#28a745')
)

# Update layout
fig.update_layout(
    title={
        'text': '<b>Perbandingan Harga Jual Berdasarkan Tipe Kendaraan</b><br>' +
                '<sub>Apakah SUV Lebih Mahal dari Sedan?</sub>',
        'y': 0.95,
        'x': 0.5,
        'xanchor': 'center',
        'yanchor': 'top',
        'font': dict(size=22, color='#1a1a1a', family='Arial Black')
    },
    xaxis=dict(
        title=dict(
            text='<b>Body Type</b>',
            font=dict(size=14, color='#2c3e50', family='Arial')
        ),
        tickfont=dict(size=12),
        showgrid=False
    ),
    yaxis=dict(
        title=dict(
            text='<b>Average Selling Price (USD)</b>',
            font=dict(size=14, color='#2c3e50', family='Arial')
        ),
        showgrid=True,
        gridcolor='rgba(200, 200, 200, 0.2)',
        tickformat='$,.0f',
        tickfont=dict(size=11),
        range=[0, body_price_main['mean_price'].max() * 1.15]
    ),
    plot_bgcolor='rgba(250, 250, 250, 1)',
    paper_bgcolor='white',
    showlegend=False,
    width=1100,
    height=650,
    margin=dict(l=80, r=80, t=120, b=120)
)

# Calculate the price difference
price_diff_suv_sedan = suv_avg - sedan_avg
percentage_diff = (price_diff_suv_sedan / sedan_avg) * 100

# Add annotation with hypothesis result
hypothesis_result = "CONFIRMED" if price_diff_suv_sedan > 0 else "REJECTED"
result_color = "#28a745" if price_diff_suv_sedan > 0 else "#dc3545"

# fig.add_annotation(
#     text=f"<b>Hypothesis Result: {hypothesis_result}</b><br>" +
#          f"SUVs sell for <b>${suv_avg:,.0f}</b> on average<br>" +
#          f"Sedans sell for <b>${sedan_avg:,.0f}</b> on average<br>" +
#          f"Difference: <b>${price_diff_suv_sedan:,.0f}</b> ({percentage_diff:.1f}% higher for SUVs)",
#     xref="paper", yref="paper",
#     x=0.5, y=-0.20,
#     showarrow=False,
#     font=dict(size=12, color='#2c3e50', family='Arial'),
#     align="center",
#     bgcolor="rgba(255, 255, 255, 0.95)",
#     borderpad=15,
#     bordercolor=result_color,
#     borderwidth=3
# )

fig.show()

# Print detailed comparison
print("\n" + "="*80)
print(" " * 25 + "HYPOTHESIS TEST RESULTS")
print("="*80)
print(f"\nHypothesis: SUVs have higher average selling prices than Sedans")
print(f"\nResult: {hypothesis_result}")
print(f"\nEvidence:")
print(f"  SUV Average Price:    ${suv_avg:>12,.2f}  (n={body_price_main[body_price_main['body_type'] == 'SUV']['count'].values[0]:,})")
print(f"  Sedan Average Price:  ${sedan_avg:>12,.2f}  (n={body_price_main[body_price_main['body_type'] == 'Sedan']['count'].values[0]:,})")
print(f"  Price Difference:     ${price_diff_suv_sedan:>12,.2f}")
print(f"  Percentage Difference: {percentage_diff:>11.2f}%")
print(f"\nConclusion:")
if price_diff_suv_sedan > 0:
    print(f"  SUVs are indeed MORE EXPENSIVE than Sedans by ${price_diff_suv_sedan:,.2f} ({percentage_diff:.1f}%)")
    print(f"  The hypothesis is CONFIRMED.")
else:
    print(f"  SUVs are actually CHEAPER than Sedans by ${abs(price_diff_suv_sedan):,.2f}")
    print(f"  The hypothesis is REJECTED.")

print(f"\nAdditional Insights:")
print(f"  - Highest priced body type: {body_price_main.iloc[0]['body_type']} (${body_price_main.iloc[0]['mean_price']:,.2f})")
print(f"  - Lowest priced body type: {body_price_main.iloc[-1]['body_type']} (${body_price_main.iloc[-1]['mean_price']:,.2f})")
print(f"  - Price range across body types: ${body_price_main['mean_price'].max() - body_price_main['mean_price'].min():,.2f}")
print("="*80)
```

```python
# Prepare data for visualization - Get top 10 colors (excluding unknown/missing)
color_counts = df_filled_no_outlier['color'].value_counts()
# Filter out "" which represents unknown/missing colors
color_counts_clean = color_counts[color_counts.index != '']
top_colors = color_counts_clean.head(10)

# Calculate percentages
total_vehicles = len(df_filled_no_outlier)
color_percentages = (top_colors / total_vehicles * 100).round(2)

# Create a color mapping for realistic color representation
color_visual_map = {
    'black': '#1a1a1a',
    'white': '#f5f5f5',
    'gray': '#808080',
    'silver': '#c0c0c0',
    'blue': '#4169e1',
    'red': '#dc143c',
    'gold': '#ffd700',
    'green': '#228b22',
    'burgundy': '#800020',
    'beige': '#f5f5dc',
    'brown': '#8b4513',
    'orange': '#ff8c00',
    'purple': '#9370db',
    'off-white': '#faf0e6',
    'yellow': '#ffff00'
}

# Get colors for the bars
bar_colors = [color_visual_map.get(color.lower(), '#6c757d') for color in top_colors.index]

# Create the visualization
fig = go.Figure()

# Add bar chart
fig.add_trace(go.Bar(
    x=top_colors.index,
    y=top_colors.values,
    marker=dict(
        color=bar_colors,
        line=dict(color='#333333', width=1.5),
        opacity=0.95
    ),
    text=[f'{count:,}<br>({pct}%)' for count, pct in zip(top_colors.values, color_percentages.values)],
    textposition='outside',
    textfont=dict(size=11, color='#2c3e50', family='Arial'),
    hovertemplate='<b>Color:</b> %{x}<br>' +
                  '<b>Vehicles:</b> %{y:,}<br>' +
                  '<b>Percentage:</b> %{customdata:.2f}%<br>' +
                  '<extra></extra>',
    customdata=color_percentages.values
))

# Update layout
fig.update_layout(
    title={
        'text': '<b>Warna Eksterior Paling Populer dalam Penjualan Kendaraan</b><br>' +
                '<sub>Distribusi 10 Warna Teratas dari Total 497,266 Kendaraan</sub>',
        'y': 0.95,
        'x': 0.5,
        'xanchor': 'center',
        'yanchor': 'top',
        'font': dict(size=22, color='#1a1a1a', family='Arial Black')
    },
    xaxis=dict(
        title=dict(
            text='<b>Exterior Color</b>',
            font=dict(size=14, color='#2c3e50', family='Arial')
        ),
        tickfont=dict(size=12),
        tickangle=-45,
        showgrid=False
    ),
    yaxis=dict(
        title=dict(
            text='<b>Number of Vehicles Sold</b>',
            font=dict(size=14, color='#2c3e50', family='Arial')
        ),
        showgrid=True,
        gridcolor='rgba(200, 200, 200, 0.2)',
        tickformat=',',
        tickfont=dict(size=11)
    ),
    plot_bgcolor='rgba(250, 250, 250, 1)',
    paper_bgcolor='white',
    showlegend=False,
    width=1100,
    height=650,
    margin=dict(l=80, r=80, t=120, b=120)
)

# Add annotation with insights
top_3_colors = ', '.join(top_colors.head(3).index.str.title())
top_3_total = top_colors.head(3).sum()
top_3_percentage = (top_3_total / total_vehicles * 100)

fig.add_annotation(
    text=f"<b>Key Insight:</b> The top 3 colors ({top_3_colors}) account for<br>" +
         f"<b>{top_3_total:,} vehicles ({top_3_percentage:.1f}%)</b> of total sales",
    xref="paper", yref="paper",
    x=0.5, y=-0.25,
    showarrow=False,
    font=dict(size=12, color='#2c3e50', family='Arial'),
    align="center",
    bgcolor="rgba(255, 255, 255, 0.95)",
    borderpad=15,
    bordercolor="#4169e1",
    borderwidth=2
)

fig.show()

# Print detailed statistics
print("\n" + "="*80)
print(" " * 22 + "MOST POPULAR EXTERIOR COLORS")
print("="*80)
print(f"\nTotal vehicles analyzed: {total_vehicles:,}")
print(f"Total unique colors: {df_filled_no_outlier['color'].nunique()}")
print(f"\nTop 10 Colors Breakdown:")
print("-"*80)
print(f"{'Rank':<6} {'Color':<15} {'Count':<15} {'Percentage':<15} {'Visual':<20}")
print("-"*80)

for rank, (color, count) in enumerate(top_colors.items(), 1):
    percentage = (count / total_vehicles * 100)
    bar_length = int(percentage / 2)  # Scale for visual representation
    visual_bar = '' * bar_length
    print(f"{rank:<6} {color.title():<15} {count:>13,} {percentage:>13.2f}% {visual_bar:<20}")

print("-"*80)
print(f"\nTop 3 Colors Summary:")
print(f"  Combined Sales: {top_3_total:,} vehicles")
print(f"  Market Share: {top_3_percentage:.2f}%")
print(f"\nColor Insights:")
print(f"  - Most popular: {top_colors.index[0].title()} ({top_colors.values[0]:,} vehicles, {color_percentages.values[0]:.2f}%)")
print(f"  - Runner-up: {top_colors.index[1].title()} ({top_colors.values[1]:,} vehicles, {color_percentages.values[1]:.2f}%)")
print(f"  - Third place: {top_colors.index[2].title()} ({top_colors.values[2]:,} vehicles, {color_percentages.values[2]:.2f}%)")
print(f"  - Neutral colors (Black, White, Gray, Silver): {(top_colors[['black', 'white', 'gray', 'silver']].sum() / total_vehicles * 100):.2f}%")
print("="*80)
```
